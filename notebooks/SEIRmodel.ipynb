   # Les modéles 
   
   On distingue plusieurs modèles mathématiques de propagation des virus qu'on présentera brièvement dans ce qui suit. On ajoutera d'avantage de détails sur le modèle choisi plus loin
   
                       1°) Le modèle Bernouillie 
                       
   Il apparait au milieu du 18$^e$ à l'époque de la variole. Il s'agit en effet du premier modèle mathématique inspiré d'une problématique médicale particulièrement épidémiologique. Cependant, il n'a pas été trop utilisé dans la littérature.

  
                      2°) Le modèle de Reed-Frost
                      
   Crée en 1918 par Lowell Reed et Wade Hampton Frost, ce modèle permet la formalisation mathématique de la transmission inter-générationelle des maladies. C'est ainsi un modèle mathématique des épidémies largement utilisé mais qui par contre n'a jamais été publié. Il s'appuit sur plusieurs hypothèses de base :
   
   
   -L'infection se transmet directement d'un sujet infecté à un sujet réceptif par un certain type de contact : le contact physique direct.
  
  
   -Un sujet réceptif du groupe, après un tel contact avec un sujet infecté au cours d'un intervalle de temps donné, présente une infection et devient infectant pour les autres uniquement pendant l'intervalle suivant, puis il devient immunisé.
   
   
   -Tout individu est soumis à une probabilité  p, définie, de venir en contact adéquat avec tout autre individu désigné du groupe pendant un intervalle de temps et cette probabilité est la même pour tous les membres du groupe.
   
   
   -Les sujets sont complètement isolés des sujets extérieurs au groupe.
   
   
   -Ces conditions restent constantes pendant l'épidémie.
   
   
   Dans ce modèle également, la population est divisée en trois groupes : Susceptibles, Infectieux,Immunisés. Cependant, avant d'aller plus loin, on remarque déja les limites que constituent les hypothèses du modèle concernant la modélisation de la propagation du virus. Ainsi, nous allons explorer les autres modèles, particulèrement les modèles comportementals
   
   
                      3°) Le modèle SIR                 
                    
   Inventé en 1923 par trois chercheurs, ils essayaient de comprendre pourquoi la grande pandémie de grippe espagnole de 1918 n'avait pas infecté toute la population. Il est actuellement utilisé par tous les recherches concernant la propagation d'un virus. Il est souvent représenté dans une population fermée, décomposée en trois entités : S (personnes susceptibles d'être infectées); I ( personnes infectées); R(personnes immunisées, c'est-à-dire les personnes retirées du groupe de la chaîne de transmission). Il est clair que reposant sur seulement 3 catégories de personnes le modèle SIR seul est inutilisable de nos jours.
   
                      4°) Le modèle SEIR 
                      
   Ce modèle est un raffinement du modèle SIR.Cependant, il prend en compte une période de latence dans le processus de transmission. Ce dernier  est souvent dû à une inoculation initiale avec un très petit nombre d'unités pathogènes. Il s'ensuit une période pendant laquelle l'agent pathogène se reproduit rapidement au sein de l'hôte. Pendant cette étape, l'abondance de l'agent pathogène est trop faible pour permettre une transmission active à d'autres hôtes sensibles, et pourtant l'agent pathogène est présent. Par conséquent, l'hôte ne peut pas être classé comme susceptible, infectieux ou guéri. Il dès lors important d'introduire une nouvelle catégorie pour ces individus qui sont infectés mais pas encore infectieux. Ces personnes sont appelées "exposées" et sont représentées par la variable E dans les modèles SEIR.
   
                 
   Notons également d'autres modèles comportementals qui sont des raffinements du modèle de base SIR. Ces modèles sont entre autres Le SIS où la guérison d'une infection est suivie d'un retour instantané dans le groupe de personnes susceptibles du fait que la maladie ne confère aucune immunité durable; il y'a également le SEIS où on rajoute le groupe des individus exposés dans le modèle de SIS.
   On voit clairement donc que le modèle de base peut etre amélioré en fonction de la réalité de l'étude. 
 
   

# Choix du modèle

   Nous avons ainsi choisi le modèle SEIR qui est plus adapté dans l'étude de la propagation du coronavirus qui s'évit actuellement dans le monde. Comme mentionné plus haut, on distingue dans ce modèle quatre entités(Voir figure) : Les susceptibles, les exposés, les infectieux et les guéris. 
   

   
   Pour de nombreuses infections importantes, il existe une période d'incubation importante pendant laquelle les individus ont été infectés mais ne sont pas encore eux-mêmes infectieux. Pendant cette période, l'individu est mis dans le groupe des personnes exposées noté E. Le fait que le nombre d'individus sensibles, infectés et guéris peut varier dans le temps (même si la taille de la population totale reste constante), nous  impose de représenter les variables en une fonction de t (temps) : S(t), I(t) et R(t).en supposant la présence d'une dynamique vitale avec le taux de mortalité  ${\displaystyle \mu }$  Nous avons le modèle suivant (On s'adataptera lors de l'application numérique) : 
    

 \begin{aligned}{\frac {dS}{dt}}&= \mu - (\beta I + \mu) S \\[8pt]{\frac {dE}{dt}}&=\beta {{S}{I}}-(\mu +a)E\\[8pt]{\frac {dI}{dt}}&=aE-(\gamma +\mu )I\\[8pt]{\frac {dR}{dt}}&=\gamma I-\mu R.\end{aligned}
 
 Avec $\mu$ qui est le taux de mortalité par habitant, et le taux de natalité au niveau de la population; $\beta$ le taux de transmission et intègre le taux de rencontre entre les individus sensibles et infectieux ainsi que la probabilité de transmission; $\gamma$ le taux d'enlèvement ou de récupération, bien que souvent nous soyons plus intéressés par sa réciprocité (1/γ) qui détermine la période infectieuse moyenne; $σ$  la vitesse à laquelle les individus passent des classes exposées aux classes infectieuses. Sa réciproque (1/σ) est la période de latence moyenne (exposée). Notons de plus que tous les taux sont indiqués en jour. Enfin, en plus de la dynamique de transmission, ce modèle permet d'utiliser des informations temporelles supplémentaires pour modéliser le taux de mortalité et la charge des soins de santé.

# Implémentation de SEIR

Pour l'apllication numérique, rappleons la définition des termes suivants :  S(t), E(t), I(t) et R(t) sont le nombre d'individus sensibles, latents, infectieux et retirés (guéris) au moment t ; la période latente est supposée identique à l'incubation et la période infectieuse moyenne est également supposée égale à l'intervalle sériel moins la période latente moyenne). 

Pour cette application, on utilisera les différentes fonctions différentielles suivantes afin de trouver S,E,I et R : 
 \begin{aligned}{\frac {dS}{dt}}&=  - \frac{R_t}{T_{inf}}IS \\[8pt]{\frac {dE}{dt}}&=\frac{R_t}{T_{inf}}IS-T_{inc}^{-1}E\\[8pt]{\frac {dI}{dt}}&=T_{inc}^{-1}E - T_{inf}^{-1}I \\[8pt]{\frac {dR}{dt}}&=T_{inf}^{-1}I.\end{aligned}
 
avec $R_t$ est le nombre de reproduction et décrit l'état où aucun individu n'est infecté ou immunisé; $T_{inf}$ la durée moyenne d'infection; $T_{inc}$ la durée moyenne d'incubation

```python
# équation des susceptibles
def dS_dt(S, I, R_t, T_inf):
    return -(R_t / T_inf) * I * S

# équation des exposés
def dE_dt(S, E, I, R_t, T_inf, T_inc):
    return (R_t / T_inf) * I * S - (T_inc**-1) * E

# équation des infectés
def dI_dt(I, E, T_inc, T_inf):
    return (T_inc**-1) * E - (T_inf**-1) * I

# equation des guéris
def dR_dt(I, T_inf):
    return (T_inf**-1) * I
```

```python
def SEIR_model(t, y, R_t, T_inf, T_inc):
    
    if callable(R_t):
        reproduction = R_t(t)
    else:
        reproduction = R_t
        
    S, E, I, R = y
    
    S_out = dS_dt(S, I, reproduction, T_inf)
    E_out = dE_dt(S, E, I, reproduction, T_inf, T_inc)
    I_out = dI_dt(I, E, T_inc, T_inf)
    R_out = dR_dt(I, T_inf)
    
    return [S_out, E_out, I_out, R_out]
```

```python

def plot_model_and_predict(data, pop, solution, title='modèle de SEIR'):
    sus, exp, inf, rec = solution.y
    
    f = plt.figure(figsize=(16,5))
    ax = f.add_subplot(1,2,1)
    #ax.plot(sus, 'b', label='Susceptible');
    ax.plot(exp, 'y', label='Exposed');
    ax.plot(inf, 'r', label='Infected');
    ax.plot(rec, 'c', label='Recovered/deceased');
    plt.title(title)
    plt.xlabel("Days", fontsize=10);
    plt.ylabel("Fraction of population", fontsize=10);
    plt.legend(loc='best');
    
    ax2 = f.add_subplot(1,2,2)
    preds = np.clip((inf + rec) * pop ,0,np.inf)
    ax2.plot(range(len(data)),preds[:len(data)],label = 'Predict Confirmed')
    ax2.plot(range(len(data)),data['Confirmed'])
    plt.title('Model predict and data')
    plt.ylabel("Population", fontsize=10);
    plt.xlabel("Days", fontsize=10);
    plt.legend(loc='best');
```

Prenons l'exemple de l'Italy et de la France. Nous allons voir l'implémentation du modèle sans intervention dans ces deux pays.

```python
Country = 'Italy'
N = 7_009_999

n_infected = Italy_df['Confirmed'].iloc[0] # début du premier cas confirmé à la première date de la base de données
max_days = len(Italy_df)# le nombre de jours que l'on veut prévoir

# Statistiques initiales pour le modèle SEIR
s = (N - n_infected)/ N
e = 0.
i = n_infected / N
r = 0.
```

Pour la définission des variables, on a utilisé les résultats des estimations des paramétres de l'épidémie de kucharski et al dans "Analysis and projections of transmission dynamics of nCoV in Wuhan". 

```python
T_inc = 5.2 
T_inf = 2.9 
R_0 = 3.954 

## Résolution du modèle 
sol = solve_ivp(SEIR_model, [0, max_days], [s, e, i, r], args=(R_0, T_inf, T_inc), 
                t_eval=np.arange(max_days))

## Plot result
plot_model_and_predict(Italy_df, N, sol, title = 'SEIR Model In Italy (without intervention)')
```

On voit que si on n'effectue pas d'intervention, le nombre de cas confirmé va continuer son augmentation.
